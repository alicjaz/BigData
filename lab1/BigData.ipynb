```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-4, 4)

f1 = 4*x**2

f2 = x**4 - 12*x**2 + 20

plt.plot(x, f1)
plt.plot(x, f2, 'r')

plt.show()
```

```python
column_names = ['mpg', 'cylinders', 'displacement [cu]',
'horsepower','weight', 'acceleration', 'year', 'origin', 'name']
```

```python
df = pd.read_csv(url, names=column_names, delim_whitespace=True)
```

```python
df = df.replace('?', np.nan)
```

```python
df
```

```python
df['horsepower'] = pd.to_numeric(df['horsepower'])
```

```python
df['full_date']=pd.to_datetime(df.year, format='%y')
```

```python
df['year'] = df['full_date'].dt.year
```

```python
df.head()
df.dtypes
```

```python
df.groupby('year')['horsepower'].mean().plot()
```

```python
df.groupby('year').agg({"horsepower":['mean'], 'acceleration':['mean'], 'mpg':["mean"]}).plot()
```

```python
df.groupby('year').agg({"horsepower":['mean'], 'acceleration':['mean'], 'mpg':["mean"]}).head()
```

```python
fig, ax = plt.subplots()
ax.scatter(x=df['mpg'], y=df['displacement [cu]'])
```

```python
import seaborn as sns
sns.scatterplot(data=df, x='mpg', y='displacement [cu]', hue='cylinders', size='weight')
```
